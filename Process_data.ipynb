```python
# Import data cần thiết cho phân tích
reviews = pd.read_csv("/Users/phamvantung/Downloads/archive (9)/scrapped_Facebook_EN.csv") 
reviews.head(10)
# Xem thông tin cơ bản về dữ liệu 
print(reviews.info())
```

```python
# Kiểm tra missing
reviews.isnull().sum()
# Kiểm tra dup
reviews.duplicated().sum()
reviews.describe()
# Loại bỏ đi những cột chưa quá nhiều giá trị null và các cột chưa có ý nghĩa
reviews = reviews.drop(columns=['userImage', 'replyContent', 'repliedAt'])
```

```python
reviews.tail(10)
```

```python
# Thực hiện xử lý cột "at" mang ý nghĩa thời gian
reviews.rename(columns={'at': 'timestamp'}, inplace=True)
# Định dạng lại cột ngày tháng
reviews['timestamp'] = pd.to_datetime(reviews['timestamp'].astype(str), format='%Y-%m-%d %H:%M:%S')
# Tạo thêm cột thứ cho thời gian
# Chuyển đổi cột 'timestamp' sang kiểu datetime
reviews['timestamp'] = pd.to_datetime(reviews['timestamp'])

# Tạo các cột 'year', 'month', 'day_of_week' và 'time_of_day'
reviews['year'] = reviews['timestamp'].dt.year
reviews['month'] = reviews['timestamp'].dt.month
reviews['day_of_week'] = reviews['timestamp'].dt.day_name()  # Hoặc bạn có thể dùng .dt.weekday nếu muốn số thứ trong tuần
reviews['time_of_day'] = reviews['timestamp'].dt.hour.apply(lambda x: 'morning' if 6 <= x < 12 else 
                                                            'afternoon' if 12 <= x < 18 else
                                                            'evening' if 18 <= x < 22 else 'night')


reviews.drop(columns=['timestamp'], inplace=True)
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Tải stopwords và wordnet
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

# Hàm xử lý văn bản
def preprocess_text(text):
    # Kiểm tra nếu text là chuỗi
    if isinstance(text, str):
        # Loại bỏ ký tự đặc biệt và chuyển về chữ thường
        text = re.sub(r'[^\w\s]', '', text.lower())
        # Tách từ
        words = word_tokenize(text)
        # Loại bỏ stop words và lemmatize
        words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
        return " ".join(words)
    else:
        # Nếu không phải chuỗi, trả về chuỗi rỗng hoặc một giá trị khác
        return ""


# Áp dụng vào cột Review
reviews['Cleaned_Review'] = reviews['content'].apply(preprocess_text)
reviews
```

```python
# Thêm cột 
from textblob import TextBlob

# Thêm một cột 'sentiment' vào DataFrame
reviews['sentiment'] = reviews['Cleaned_Review'].apply(lambda x: TextBlob(x).sentiment.polarity)
reviews['sentiment_label'] = reviews['sentiment'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))

```

```python
reviews = reviews.drop(columns= ['userName', 'reviewCreatedVersion', 'appVersion', 'sentiment'])
```

```python
month_mapping = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr',
    5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug',
    9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}

# Đổi số thành tên tháng viết tắt
reviews['Month'] = reviews['month'].map(month_mapping)
```
