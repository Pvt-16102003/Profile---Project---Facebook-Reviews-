```python
# Import data c·∫ßn thi·∫øt cho ph√¢n t√≠ch
reviews = pd.read_csv("/Users/phamvantung/Downloads/archive (9)/scrapped_Facebook_EN.csv") 
reviews.head(10)
# Xem th√¥ng tin c∆° b·∫£n v·ªÅ d·ªØ li·ªáu 
print(reviews.info())
```

```python
# Ki·ªÉm tra missing
reviews.isnull().sum()
# Ki·ªÉm tra dup
reviews.duplicated().sum()
reviews.describe()
# Lo·∫°i b·ªè ƒëi nh·ªØng c·ªôt ch∆∞a qu√° nhi·ªÅu gi√° tr·ªã null v√† c√°c c·ªôt ch∆∞a c√≥ √Ω nghƒ©a
reviews = reviews.drop(columns=['userImage', 'replyContent', 'repliedAt'])
```

```python
# Th·ª±c hi·ªán x·ª≠ l√Ω c·ªôt "at" mang √Ω nghƒ©a th·ªùi gian
reviews.rename(columns={'at': 'timestamp'}, inplace=True)
# ƒê·ªãnh d·∫°ng l·∫°i c·ªôt ng√†y th√°ng
reviews['timestamp'] = pd.to_datetime(reviews['timestamp'].astype(str), format='%Y-%m-%d %H:%M:%S')
# T·∫°o th√™m c·ªôt th·ª© cho th·ªùi gian
# Chuy·ªÉn ƒë·ªïi c·ªôt 'timestamp' sang ki·ªÉu datetime
reviews['timestamp'] = pd.to_datetime(reviews['timestamp'])

# T·∫°o c√°c c·ªôt 'year', 'month', 'day_of_week' v√† 'time_of_day'
reviews['year'] = reviews['timestamp'].dt.year
reviews['month'] = reviews['timestamp'].dt.month
reviews['day_of_week'] = reviews['timestamp'].dt.day_name()  # Ho·∫∑c b·∫°n c√≥ th·ªÉ d√πng .dt.weekday n·∫øu mu·ªën s·ªë th·ª© trong tu·∫ßn
reviews['time_of_day'] = reviews['timestamp'].dt.hour.apply(lambda x: 'morning' if 6 <= x < 12 else 
                                                            'afternoon' if 12 <= x < 18 else
                                                            'evening' if 18 <= x < 22 else 'night')


reviews.drop(columns=['timestamp'], inplace=True)
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# T·∫£i stopwords v√† wordnet
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

# H√†m x·ª≠ l√Ω vƒÉn b·∫£n
def preprocess_text(text):
    # Ki·ªÉm tra n·∫øu text l√† chu·ªói
    if isinstance(text, str):
        # Lo·∫°i b·ªè k√Ω t·ª± ƒë·∫∑c bi·ªát v√† chuy·ªÉn v·ªÅ ch·ªØ th∆∞·ªùng
        text = re.sub(r'[^\w\s]', '', text.lower())
        # T√°ch t·ª´
        words = word_tokenize(text)
        # Lo·∫°i b·ªè stop words v√† lemmatize
        words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
        return " ".join(words)
    else:
        # N·∫øu kh√¥ng ph·∫£i chu·ªói, tr·∫£ v·ªÅ chu·ªói r·ªóng ho·∫∑c m·ªôt gi√° tr·ªã kh√°c
        return ""


# √Åp d·ª•ng v√†o c·ªôt Review
reviews['Cleaned_Review'] = reviews['content'].apply(preprocess_text)
reviews
```

```python
# Th√™m c·ªôt 
from textblob import TextBlob

# Th√™m m·ªôt c·ªôt 'sentiment' v√†o DataFrame
reviews['sentiment'] = reviews['Cleaned_Review'].apply(lambda x: TextBlob(x).sentiment.polarity)
reviews['sentiment_label'] = reviews['sentiment'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))

```

```python
reviews = reviews.drop(columns= ['userName', 'reviewCreatedVersion', 'appVersion', 'sentiment'])
```
